```python
# Abrindo OpenCV
import cv2
import common
import numpy as np

%matplotlib inline 
from matplotlib import pyplot as plt
import pylab
pylab.rcParams['figure.figsize'] = (20.0, 20.0) 
```

```python
# Lendo a imagem
ruido1 = cv2.imread('foto_ruido.png')
plt.imshow(ruido1)

filtro = cv2.medianBlur(ruido1,5)

# Nesse código a imagem bruta será mostrada em cima e a imagem tratada
# será mostrada em baixo, a diferença é enorme.
plt.subplot(121),plt.imshow(ruido1),plt.title('Imagem Bruta (Com Ruido)')
plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(122),plt.imshow(filtro),plt.title('Imagem Tratada')
plt.xticks([]), plt.yticks([])
plt.show()
print("\n\tDa a impressão que a imagem foi limpa")
```
